## CMAE + DQN + HER Analysis

```python
DQNHERLoss01 = pd.read_csv("log/CMAE/CMAE_PUSH_BOX_HER_lbda0_p1_0losses.csv")
DQNHERLoss01 = DQNHERLoss01[(np.abs(stats.zscore(DQNHERLoss01)) < 3).all(axis=1)]
axes1 = DQNHERLoss01[["q_learner1_loss", "q_learner2_loss"]].plot.line(subplots=True, figsize=(30, 15), xlabel="Episode", ylabel="loss")
axes1[0].set_title('CMAE + DQN + HER, Lambda = 0, p = 1', fontsize=18)

DQNHERLoss02 = pd.read_csv("log/CMAE/CMAE_PUSH_BOX_HER_lbda0_p2_0losses.csv")
DQNHERLoss02 = DQNHERLoss02[(np.abs(stats.zscore(DQNHERLoss02)) < 3).all(axis=1)]
axes2 = DQNHERLoss02[["q_learner1_loss", "q_learner2_loss"]].plot.line(subplots=True, figsize=(30, 15), xlabel="Episode", ylabel="loss")
axes2[0].set_title('CMAE + DQN + HER, Lambda = 0, p = 2', fontsize=18)


DQNHERLoss11 = pd.read_csv("log/CMAE/CMAE_PUSH_BOX_HER_lbda1_p1_0losses.csv")
DQNHERLoss11 = DQNHERLoss11[(np.abs(stats.zscore(DQNHERLoss11)) < 3).all(axis=1)]
axes3 = DQNHERLoss11[["q_learner1_loss", "q_learner2_loss"]].plot.line(subplots=True, figsize=(30, 15), xlabel="Episode", ylabel="loss")
axes3[0].set_title('CMAE + DQN + HER, Lambda = 1, p = 1', fontsize=18)

DQNHERLoss12 = pd.read_csv("log/CMAE/CMAE_PUSH_BOX_HER_lbda1_p2_0losses.csv")
DQNHERLoss12 = DQNHERLoss12[(np.abs(stats.zscore(DQNHERLoss12)) < 3).all(axis=1)]
axes4 = DQNHERLoss12[["q_learner1_loss", "q_learner2_loss"]].plot.line(subplots=True, figsize=(30, 15), xlabel="Episode", ylabel="loss")
axes4[0].set_title('CMAE + DQN + HER, Lambda = 1, p = 2', fontsize=18)



display(DQNHERLoss01.describe())
display(DQNHERLoss02.describe())
display(DQNHERLoss11.describe())
display(DQNHERLoss12.describe())
```

```python
def plotEvalReward(logPath, lbda, p):
    dfPushBox = pd.read_csv(logPath).T
    new_header = dfPushBox.iloc[0]
    dfPushBox = dfPushBox[1:]
    dfPushBox.columns = new_header
    ax = dfPushBox[['train_rew', 'eval_rew', 'step']].plot.line(x='step')
    ax.set_xlabel("# of Env. Steps")
    ax.set_ylabel("Reward")
    ax.set_title(f"CMAE + DQN + HER, Lambda={lbda}, p={p}")
```

```python
plotEvalReward("log/CMAE/CMAE_PUSH_BOX_HER_lbda0_p1_0/data.csv", 0, 1)
```

```python
plotEvalReward("log/CMAE/CMAE_PUSH_BOX_HER_lbda0_p2_0/data.csv", 0, 2)
```

```python
plotEvalReward("log/CMAE/CMAE_PUSH_BOX_HER_lbda1_p1_0/data.csv", 1, 1)
```

```python
plotEvalReward("log/CMAE/CMAE_PUSH_BOX_HER_lbda1_p2_0/data.csv", 1, 2)
```

# PPO

```python
PPOResults = pd.read_csv("ppo_results.csv", names=['iteration', 'episode reward max', 'policy loss', 'number of steps taken', 'reached goal'], header=None)

############### Plot loss ###############
PPOResults[['policy loss']].plot(subplots=True)

############### Find Frequency of reaching goal ###############
reached_goal = PPOResults['reached goal']
print(np.unique(reached_goal))
figure(num=None, figsize=(20,10), dpi=80, facecolor='w', edgecolor='k')
plt.scatter(x = reached_goal.index, y= reached_goal, color='g', marker='d')
```
